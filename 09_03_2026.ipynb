```python
# Loan Eligibily Checker

#original proof docs (id proof, address proof),
#allotment based on age factor, cibil score,
#income verification,
#risk level,
#assurance of gurantee person,
#history of previous loans and payments,
#emplyement verification letter,
#loan amount eligible

# Applicant Details
cust_age = 20
cust_cibil = 400
cust_incomeproof = True
cust_risklevel = "Low"
cust_guarantee = True
cust_history_loans = False
cust_emp_letter = True
cust_amount_avail = 120000

# Eligibility Criteria
min_age = 18
min_cibil = 300
max_cibil = 900
min_risklevel = "Low"
max_risklevel = "Medium"
min_loan_amount_threshold = 100000

# --- Loan Eligibility Logic ---
is_eligible = (
    cust_age >= min_age and
    min_cibil <= cust_cibil <= max_cibil and
    cust_incomeproof and
    (cust_risklevel == min_risklevel or cust_risklevel == max_risklevel) and
    cust_guarantee and
    not cust_history_loans and
    cust_emp_letter and
    cust_amount_avail >= min_loan_amount_threshold
)

print("\n--- Loan Application Status ---")
if is_eligible:
    print("Status: APPROVED")
    print(f"Eligible for a loan of up to: ₹{cust_amount_avail:,.2f}")
else:
    print("Status: REJECTED")
    print("Note: Detailed reasons for rejection would typically be provided in a more comprehensive system.")
```
