```python
# Given sets
account_holders = {"C101", "C102", "C103", "C104", "C105", "C106"}
kyc_verified = {"C102", "C104", "C106", "C108"}
defaulters = {"C107", "C105"}

# Create set of customers eligible for loan
eligible_customers = account_holders & kyc_verified - defaulters

print("Eligible customers:", eligible_customers)

# Display total eligible customers
print("Total eligible customers:", len(eligible_customers))

# Check whether C104 is eligible
print("Is C104 eligible?", "C104" in eligible_customers)

# Customers either verified or defaulters but not both
result = kyc_verified ^ defaulters
print("Verified or Defaulters but not both:", result)

# Update operations
kyc_verified.add("C107") # add C107

account_holders.discard("C105") # remove safely

print("Updated KYC verified:", kyc_verified)
print("Updated account holders:", account_holders)

# Clean up
backup = kyc_verified.copy()

kyc_verified.clear()

print("Backup KYC:", backup)
print("Original KYC after clear:", kyc_verified)
```

```python
# Given sets
savings_customers = {"C101","C102","C103","C104","C105"}
loan_customers = {"C104","C105","C106","C107"}
credit_card_customers = {"C102","C105","C108"}

# 1. Unique customers using at least one service (Union)
all_customers = savings_customers | loan_customers | credit_card_customers
print("All customers:", all_customers)

# 2. Customers having both savings and loan
both_savings_loan = savings_customers & loan_customers
print("Savings and Loan:", both_savings_loan)

# 3. Customers with loan but not savings
loan_not_savings = loan_customers - savings_customers
print("Loan but not Savings:", loan_not_savings)

# 4. Customers with only savings or credit card (not both)
only_one = savings_customers ^ credit_card_customers
print("Savings or Credit card but not both:", only_one)

# 5. Check if credit card customers are subset of savings
print("Credit card subset of savings:", credit_card_customers.issubset(savings_customers))

# 6. Check if loan and credit card customers are disjoint
print("Loan and Credit card disjoint:", loan_customers.isdisjoint(credit_card_customers))

# 7. Check if savings customers include all loan customers (superset)
print("Savings superset of loan:", savings_customers.issuperset(loan_customers))

# 8. Remove credit card customers from savings (in-place update)
savings_customers.difference_update(credit_card_customers)
print("Savings after removal:", savings_customers)
```

```python
vehicle_number = input("Enter vehicle number: ")
hours = int(input("Enter number of hours parked: "))

# Calculate fee
if hours <= 2:
    fee = 20
elif hours <= 5:
    fee = 30
else:
    fee = 50

# Store in dictionary
parking_details = {
    "Vehicle Number": vehicle_number,
    "Hours Parked": hours,
    "Total Parking Fee": "₹" + str(fee)
}

# Display result
print(parking_details)

parking_data = []

n = int(input("Enter number of vehicles: "))

for i in range(n):
    vehicle_number = input("Enter Vehicle Number: ")
    hours_parked = int(input("Enter Hours Parked: "))
    vehicle_model = input("Enter Vehicle Model: ")
    wheeler_type = input("Enter Wheeler Type (2/4): ")
    payment_category = input("Enter Payment Category: ")
    owner_name = input("Enter Owner Name: ")
    owner_cnumber = input("Enter Owner Contact Number: ")

    if hours_parked <= 2:
        price_per_hour = 20
    elif hours_parked <= 5:
        price_per_hour = 30
    else:
        price_per_hour = 50

    vehicle_set = {
        vehicle_number,
        hours_parked,
        vehicle_model,
        wheeler_type,
        payment_category,
        owner_name,
        owner_cnumber,
        price_per_hour
    }

    parking_data.append(vehicle_set)

print("\nParking Details:")
for vehicle in parking_data:
    print(vehicle)
```
