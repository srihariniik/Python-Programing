```python
# Flight Name : Number of seats left
flights = {
    "AI202": 10,
    "6E501": 5
}

# This list will store the names of people who have a ticket
passengers = []


def show_status():
    print("\n--- Current Seat Availability ---")
    for name, count in flights.items():
        print(f"Flight {name}: {count} seats left")
    print(f"Booked Passengers: {passengers}")

def book_ticket():
    name = input("Enter Passenger Name: ")
    f_id = input("Enter Flight ID (AI202 or 6E501): ")

    if f_id in flights:
        if flights[f_id] > 0:
            flights[f_id] = flights[f_id] - 1  # Reduce seat count
            passengers.append(name)             # Add name to our list
            print(f"Success! {name} is booked.")
        else:
            print("Sorry, no seats left!")
    else:
        print("Flight not found.")

def cancel_ticket():
    name = input("Enter Passenger Name to cancel: ")
    f_id = input("Enter Flight ID: ")

    # Check if the person is actually in our list
    if name in passengers:
        passengers.remove(name)            # Remove name from list
        flights[f_id] = flights[f_id] + 1  # Add the seat back
        print(f"Success! Ticket for {name} has been cancelled.")
    else:
        print("Passenger name not found in our records.")

# 3. THE MENU (The "Interface")
while True:
    print("\n--- AIRLINE SYSTEM ---")
    print("1. View Flights")
    print("2. Book Ticket")
    print("3. Cancel Ticket")
    print("4. Exit")

    choice = input("Select an option (1-4): ")

    if choice == "1":
        show_status()
    elif choice == "2":
        book_ticket()
    elif choice == "3":
        cancel_ticket()
    elif choice == "4":
        print("Closing System...")
        break
    else:
        print("Invalid choice, try again.")
```
